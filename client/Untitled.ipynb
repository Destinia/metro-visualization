```python
import csv
import json
rows = open('./src/utils/travalTime.txt').read().splitlines()

split = [row.split('\t') for row in rows]
stop_time = {row[1][2:-1]: int(row[5]) for row in split if row[5] != 0}
travel_time = {}
for row in split:
    travel_time[row[1][2:-1] + row[2][2:-1]] = int(row[3])
    travel_time[row[2][2:-1] + row[1][2:-1]] = int(row[3])
print(travel_time)
# json.dump(stop_time, open('./src/utils/stopTime.json', 'w'), ensure_ascii=False)
json.dump(travel_time, open('./src/utils/travelTime.json', 'w'), ensure_ascii=False)
```
